## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
y_df.head()
```

```python
# Create a list of at least 10 column names to use as X data
selected_cols = ['Education','Age', 'DistanceFromHome', 'JobSatisfaction', 'OverTime', 'StockOptionLevel',
                 'WorkLifeBalance', 'YearsAtCompany','YearsSinceLastPromotion', 'NumCompaniesWorked']
print(f"Number of Features: {len(selected_cols)}")

# Create X_df using your selected columns
X_df = attrition_df[selected_cols]
display(X_df.head(3))
# Show the data types for X_df
display(X_df.dtypes)
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X_df, y_df)
```

```python
X_train.head()
```

```python

# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
X_train['OverTime'] = X_train['OverTime'].astype(str)
X_test['OverTime'] = X_test['OverTime'].astype(str)

X_train['OverTime'] = X_train['OverTime'].map({'Yes': 1, 'No': 0})
X_test['OverTime'] = X_test['OverTime'].map({'Yes': 1, 'No': 0})

X_train.dtypes
```

```python
X_train.head()
```

```python
X_train['OverTime'].unique()
```

```python
X_df.columns
```

```python
# Create a StandardScaler
scaler = StandardScaler()
cols_to_scale = ['Education', 'Age', 'DistanceFromHome', 'JobSatisfaction',
       'StockOptionLevel', 'WorkLifeBalance', 'YearsAtCompany',
       'YearsSinceLastPromotion', 'NumCompaniesWorked','OverTime']

X_scaler = scaler.fit(X_train)

# Fit the StandardScaler to the training data
X_train_scaled =X_scaler.transform(X_train)

# Scale the training and testing data
X_test_scaled = X_scaler.transform(X_test)

# Display preview
#display(X_train_scaled.head(3))
#display(X_test_scaled.head(3))
```

```python
# Create a OneHotEncoder for the Department column
from sklearn.preprocessing import OneHotEncoder
dep_ohe = OneHotEncoder(sparse_output=False,
                        handle_unknown='ignore',
                        drop = None)

# Fit the encoder to the training data
dep_ohe.fit(np.array(y_train['Department']).reshape(-1,1))

# Create two new variables by applying the encoder
# to the training and testing data
dep_train_encoded = dep_ohe.transform(np.array(y_train['Department']).reshape(-1,1))
dep_test_encoded = dep_ohe.transform(np.array(y_test['Department']).reshape(-1,1))

```

```python
# Create a OneHotEncoder for the Attrition column
att_ohe = OneHotEncoder(sparse_output=False,
                        handle_unknown='ignore',
                        drop = None)

# Fit the encoder to the training data
att_ohe.fit(np.array(y_train['Attrition']).reshape(-1,1))

# Create two new variables by applying the encoder
# to the training and testing data
att_train_encoded = att_ohe.transform(np.array(y_train['Attrition']).reshape(-1,1))
att_test_encoded = att_ohe.transform(np.array(y_test['Attrition']).reshape(-1,1))
```

```python
y_train['Department'].unique()
```

## Create, Compile, and Train the Model

```python
num_inputs = len(X_train.columns)
num_inputs
```

```python
# Find the number of columns in the X training data
num_inputs = len(X_train.columns)

# Create the input layer
input_layer = layers.Input(shape = (num_inputs,), name='input_layer')

# Create at least two shared layers
shared_layer_1 = layers.Dense(32, activation='relu', name = "shared_layer_1")(input_layer)
shared_layer_2 = layers.Dense(64, activation='relu',name = "shared_layer_2")(shared_layer_1)
shared_layer_3 = layers.Dense(32, activation='relu',name = "shared_layer_3")(shared_layer_2)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
department_dense  = layers.Dense(32, activation='relu', name = "department_dense")(shared_layer_3)

# Create the output layer
department_output = layers.Dense(len(y_train['Department'].unique()), activation='softmax', name = 'department_output')(department_dense)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
attrition_dense = layers.Dense(32, activation='relu', name ='attrition_dense')(shared_layer_3)

# Create the output layer
attrition_output = layers.Dense(len(y_train['Attrition'].unique()), activation='sigmoid', name = 'attrition_output')(attrition_dense)
```

```python
# Create the model
model = Model(inputs = input_layer, outputs = [attrition_output, department_output])

# Compile the model
model.compile(optimizer='adam',
              loss = {'attrition_output': 'binary_crossentropy',
                      'department_output': 'categorical_crossentropy'},
              metrics = {'attrition_output': 'accuracy',
                         'department_output': 'accuracy'})

# Summarize the model
model.summary()
```

```python
# Train the model
model.fit(X_train_scaled,
          {
              'attrition_output': att_train_encoded,
              'department_output': dep_train_encoded
          },
          epochs = 100,
          batch_size = 32,
          validation_split = 0.2
          )

```

```python
# Evaluate the model with the testing data
results =model.evaluate(np.array(X_test_scaled), {
    'attrition_output': att_test_encoded,
    'department_output': dep_test_encoded
})
```
